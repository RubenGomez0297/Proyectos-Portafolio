```python
#Importamos las libreria a utilizar
import pandas as pd
import numpy as np
```

```python
#importamos el csv con el que trabajaremos. En este caso usamos la opcion "encoding" debido al formato con el cual se descargo el csv de kaggle
df=pd.read_csv("Sample - Superstore.csv", encoding='latin-1')
df
```

```python
#Validamos los tipos de atributos que tiene el df
df.dtypes
```

```python
#validamos si hay valores nulos en cada columna
missing_data = df.isnull()
for column in missing_data.columns.values.tolist():
    print (missing_data[column].value_counts())
    print("")    
```

```python
#En este caso no hay nulos
#procedemos a dividir el dataframe en las tablas que tenemos en nuestra base de datos SUPERSTORE (PRODUCTS, CLIENT, SALES)

df_products=df[["Product ID","Category","Sub-Category","Product Name"]]

df_client=df[["Customer ID","Customer Name","Segment","Country","City","State","Postal Code","Region"]]

df_sales=df
```

```python
#Limpiamos de duplicados el df_productos, ya que esta sera una tabla maestra de productos
df_products_sin  = df_products.drop_duplicates(subset=['Product ID'], keep='first').reset_index(drop=True)
df_products_sin
```

```python
# Cargar datos limpios a PRODCUTS
import pyodbc

conn = pyodbc.connect(
    "Driver={ODBC Driver 17 for SQL Server};"
    f"Server=DESKTOP-EO0JDO7\\MSSQLSERVER_RGD;"
    f"Database=SUPERSTORE;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()

# Insertar datos con executemany
data = [tuple(x) for x in df_products_sin.to_numpy()]
cursor.fast_executemany = True  
cursor.executemany(
    "INSERT INTO PRODUCTS ([product_id], category, [sub_category], [prodcut_name]) VALUES (?, ?, ?, ?)",
    data    
)
conn.commit()
```

```python
#Limpiamos de duplicados el df_client, ya que esta sera una tabla maestra de clientes
df_client_sin  = df_client.drop_duplicates(subset=['Customer ID'], keep='first').reset_index(drop=True)
df_client_sin
```

```python
# Cargar datos limpios a CLIENT
import pyodbc

conn = pyodbc.connect(
    "Driver={ODBC Driver 17 for SQL Server};"
    f"Server=DESKTOP-EO0JDO7\\MSSQLSERVER_RGD;"
    f"Database=SUPERSTORE;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()

# Insertar datos con executemany
data = [tuple(x) for x in df_client_sin.to_numpy()]
cursor.fast_executemany = True  
cursor.executemany(
    "INSERT INTO CLIENT ([customer_ID], [customer_name], [segment], [country],[city],[state_name],[postal_code],[region]) VALUES (?, ?, ?, ?, ?, ?, ?, ?)",
    data    
)
conn.commit()
```

```python
#Eliminamos la columna Row ID ya que la crearemos directamente en sql server como pk al ingresar un nuevo registro
df_sales = df_sales.drop(columns=['Row ID'])
df_sales
```

```python
#damos formato a las campos de fechas, y validamos el cambio 
df_sales['Order Date'] = pd.to_datetime(df_sales['Order Date'], errors='coerce')
df_sales['Ship Date'] = pd.to_datetime(df_sales['Ship Date'], errors='coerce')
df_sales.dtypes
```

```python
#Ahora cargamos la tabla SALES, esta contiene toda la info con respecto a las ordenes de compra
import pyodbc

conn = pyodbc.connect(
    "Driver={ODBC Driver 17 for SQL Server};"
    f"Server=DESKTOP-EO0JDO7\\MSSQLSERVER_RGD;"
    f"Database=SUPERSTORE;"
    "Trusted_Connection=yes;"
)

cursor = conn.cursor()

# Insertar datos con executemany
data = [tuple(x) for x in df_sales.to_numpy()]
cursor.fast_executemany = True  
cursor.executemany(
    "INSERT INTO SALES ([order_id], [order_date], [ship_date], [ship_mode],[customer_id],[customer_name],[segment],[country],[city],[state_name],[postal_code],[region],[product_id],[category],[sub_category],[product_name],[sales],[quantity],[discount],[profit]) VALUES (?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?, ?)",
    data    
)
conn.commit()
```
